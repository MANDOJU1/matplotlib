```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
```

```python
for fpath in fm.findSystemFonts(fontpaths=None, fontext='ttf'):
  fm.fontManager.addfont(fpath)

plt.rc("font",family='NanumBarunGothic')
```

### 1) 막대차트(기본)

```python
values = [175, 187, 184]
labels = ["강백호","서태웅","정대만"]

plt.bar(labels, values);
```

><b>색상</b>

```python
plt.bar(labels, values, color="r");
```

```python
plt.bar(labels, values, color=["r","g","b"]);
```

```python
plt.bar(labels, values, color=["r","g","b"], alpha=0.5);
```

><b>축 레이블 조정</b>

```python
plt.bar(labels, values);
# y 축 범위 설정
plt.ylim(175, 190)
```

><b>막대 너비 조정</b>

```python
plt.bar(labels, values, width=0.4)
```

><b>축 레이블 rotation 설정</b>

```python
# 데이터 이름 각도 바꾸기
plt.bar(labels, values, width=0.4)
plt.xticks(rotation=45)
plt.yticks(rotation=45);
```

><b>막대 </b>

```python
# 이름을 실명이 아닌 1번학생, 2번학생, 3번학생으로 나오게하기

plt.bar(labels, values, width=0.4)
xticks = ["1번학생","2번학생","3번학생"]
plt.xticks(labels, xticks)
```

### 2) 막대차트(심화)

**수평막대**

```python
plt.barh(labels, values);
```

```python
plt.barh(labels, values);
plt.xlim(170,195)
```

**막대 패턴**(https://matplotlib.org/stable/gallery/shapes_and_collections/hatch_style_reference.html)

```python
bar = plt.bar(labels, values);
bar[0].set_hatch("/")
bar[1].set_hatch("x")
bar[2].set_hatch("*")
```

```python
bar = plt.bar(labels, values);
plt.ylim(170, 195)

for idx, rect in enumerate(bar):
  plt.text(idx, rect.get_height() + 0.5, values[idx], ha="center")
```
